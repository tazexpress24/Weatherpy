```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time

api_key = '174460e19c7cf006b3c2e751ce3ba6fd'

# Import API key
import api_keys

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []
weather_data = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)

```

## Perform API Calls

```python
# OpenWeatherMap API Key
api_key = api_keys.api_key

# Starting URL for Weather Map API Call
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + api_key 

#Creating Empty Datframe
city_data = pd.DataFrame()

   

i = 1
set = 1
for city in cities:
    if(i == 50):
        set =  set + 1
        i = 0
        time.sleep(60)
    #response = requests.get(url).json()
    
    #city_data_df = city_data.append(city_data, ignore_index=True)   
    
    #else:
    #   citiesNotFoundinAPI.append(city)
     # Search for city       
    city_url = url +"&q=" + city
    
    #Request City Weather
    city_weather = requests.get(city_url).json()    
    if(city_weather["cod"] == 200):
        i = i + 1
        print(city_url)

#extracting data - parsing the data of max temp, humidity, latitude, Cloudiness,Windspeed,
        city_temp_max = city_weather['main']['temp_max']
        city_lat = city_weather['coord']['lat']
        city_humidity = city_weather['main']['humidity']
        city_cloudiness = city_weather['clouds']['all']
        city_windspeed = city_weather['wind']['speed']
        city_name = city_weather['name']
        city_country = city_weather['sys']['country']
        city_lng = city_weather['coord']['lon']
        city_dt = city_weather['dt']
        #print(city_name,city_cloudiness)
        #, city_country,city_date,
              #city_humidity, city_lat, city_lng, city_temp_max,
              #city_windspeed)

#create a dictionary for the variables
        weather_dict = {
        "City": city_name,
        "Cloudiness": city_cloudiness,
        "Country": city_country,
        "Date": city_dt,
        "Humidity": city_humidity,
        "Lat": city_lat,
        "Lng": city_lng,
        "Max Temp": city_temp_max,
        "Wind Speed": city_windspeed}
    
        
    
#Append the dictionary to the list,it ends up being a list of dictionaries
        weather_data.append(weather_dict)


#Create a dataframe
weather_data_pd = pd.DataFrame(weather_data)
weather_data_pd.head()








     
        
        
    
```

# City Latitude vs Max Temperature


```python
plt.scatter(weather_data_pd["Lat"],weather_data_pd["Max Temp"])
plt.xlabel("Latitude")
plt.ylabel("Max Temperature")
plt.title("City Latitude vs Max Temperature")
plt.xlim(0, 100)
plt.grid()
plt.show()
```

# Humidity (%) vs. Latitude

```python
plt.scatter(weather_data_pd["Lat"],weather_data_pd["Humidity"])
plt.xlabel("Latitude")
plt.ylabel("Humidity")
plt.title("City Latitude vs Humidity")
plt.ylim(0, 120)
plt.xlim(0, 100)
plt.grid()
plt.show()
```
